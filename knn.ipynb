```python
df=pd.read_excel('survey_dataset.xls')
```

```python

def quali_to_numeric(x):
    if x=='SSLC':
        return 1
    if x=='Plus Two':
        return 2
    if x=='Graduation':
        return 3
    if x=='PG':
        return 4
    else :
        return 0
    
def country_to_numeric(x):
    if x=='America':
        return 5
    else:
        return 3
```

```python
df['Qualification'] = df['Qualification'].apply(quali_to_numeric)
df['Country'] = df['Country'].apply(country_to_numeric)
df
```
